```python
#Problem 1, Part (b) and (c)

def make_orthonormal_vectors(v1,v2,v3):
    v1 = v1/v1.norm()
    v2 = v2 - (v2.dot_product(v1))*v2
    v2 = v2/v2.norm()
    v3 = v3 - (v3.dot_product(v1))*v1 - (v3.dot_product(v2))*v2
    v3 = v3/v3.norm()
    
    return v1,v2,v3
```
