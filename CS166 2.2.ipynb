```python
# Initial Simulation and Question 11.5

import matplotlib
matplotlib.use('TkAgg')
import pylab
from pylab import *

n = 100 # size of space: n x n
p = 0.27 # probability of initially panicky individuals

def xrange(x):

    return iter(range(x))

from numpy import zeros

def initialize():
    global config, nextconfig, density
    density = []
    number_of_ones = 0
    config = zeros([n, n])
    for x in xrange(n):
        for y in xrange(n):
            config[x, y] = 1 if random() < p else 0
            if config[x,y] == 1:
                number_of_ones += 1
    density.append(number_of_ones/10000)
    nextconfig = zeros([n, n])

def observe():
    global config, nextconfig
    cla()
    pylab.subplot(1,2,1)
    imshow(config, vmin = 0, vmax = 1, cmap = cm.binary)
    pylab.subplot(1,2,2)
    matplotlib.pyplot.plot(range(len(density)),density)
    
def update():
    global config, nextconfig, density
    number_of_ones = 0
    for x in xrange(n):
        for y in xrange(n):
            count = 0
            for dx in [-1, 0, 1]:
                for dy in [-1, 0, 1]:
                    count += config[(x + dx) % n, (y + dy) % n]
            
            nextconfig[x, y] = 1 if count >= 4 else 0
            if nextconfig[x,y] == 1:
                number_of_ones += 1
    density.append(number_of_ones/10000)
    config, nextconfig = nextconfig, config

import pycxsimulator
pycxsimulator.GUI().start(func=[initialize, observe, update])
```

```python
import matplotlib
matplotlib.use('TkAgg')
import pylab
from pylab import *

n = 100 # size of space: n x n
p = 0.4

from numpy import zeros

def xrange(x):

    return iter(range(x))


def initialize2():
    global config, nextconfig, density
    density = []
    number_of_ones = 0
    config = zeros([n, n])
    for x in xrange(n):
        for y in xrange(n):
            config[x, y] = 1 if random() < p else 0
            if config[x,y] == 1:
                number_of_ones += 1
    density.append(number_of_ones/10000)
    nextconfig = zeros([n, n])

def observe2():
    global config, nextconfig
    cla()
    pylab.subplot(1,2,1)
    imshow(config, vmin = 0, vmax = 1, cmap = cm.binary)
    pylab.subplot(1,2,2)
    matplotlib.pyplot.plot(range(len(density)),density)

        
def update2():
    global config, nextconfig, density
    number_of_ones = 0
    for x in xrange(n):
        for y in xrange(n):
            count = 0
            for dx in [-1, 0, 1]:
                for dy in [-1, 0, 1]:
                    count += config[(x + dx) % n, (y + dy) % n]
            
            nextconfig[x, y] = 1 if count >= 2 and count <= 5 else 0
            if nextconfig[x,y] == 1:
                number_of_ones += 1
    density.append(number_of_ones)
    config, nextconfig = nextconfig, config

import pycxsimulator
pycxsimulator.GUI().start(func=[initialize2, observe2, update2])
```
